## Unsupervised Data Analysis


Welcome to the "Unsupervised Data Analysis" practical session.  

Programming Language : Python 
Editor : Jupyter Notebook. 


Pandas : Data loading, processing, transformation and manipulation.
Scikit-learn : Example data source, ML and statistical analysis

This example illustrates how to apply different preprocessing and feature
imputation pipelines to different subsets of features, using
SimpleImputer, KNNImputer. This is particularly handy for the
case of datasets that contain heterogeneous data types, since we may want to
impute the numeric as well as categorical features

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE

# Load the diabetes dataset
data = load_diabetes()
df = pd.DataFrame(data=data.data, columns=data.feature_names)

# Standardize features to have zero mean and unit variance
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# Colors and labels for the legend
colors = ['r', 'g', 'b']
labels = ['Cluster 1', 'Cluster 2', 'Cluster 3']

# K-Means Clustering
kmeans = KMeans(n_clusters=2)  # Changed to 2 clusters as this dataset has two classes (diabetic or not)
kmeans_labels = kmeans.fit_predict(df_scaled)

# Visualize K-Means Clustering
plt.figure(figsize=(8, 5))
for i in range(2):  # Changed to 2 clusters
    plt.scatter(df_scaled[kmeans_labels == i, 0], df_scaled[kmeans_labels == i, 1], color=colors[i], label=labels[i])
plt.title('K-Means Clustering')
plt.legend()
plt.show()

# Hierarchical Clustering
hier_clust = AgglomerativeClustering(n_clusters=2)  # Changed to 2 clusters
hier_clust_labels = hier_clust.fit_predict(df_scaled)

# Visualize Hierarchical Clustering
plt.figure(figsize=(8, 5))
for i in range(2):  # Changed to 2 clusters
    plt.scatter(df_scaled[hier_clust_labels == i, 0], df_scaled[hier_clust_labels == i, 1], color=colors[i], label=labels[i])
plt.title('Hierarchical Clustering')
plt.legend()
plt.show()

# Expectation-Maximization (EM) Clustering using GMM
gmm = GaussianMixture(n_components=2)  # Changed to 2 clusters
gmm_labels = gmm.fit_predict(df_scaled)

# Visualize GMM
plt.figure(figsize=(8, 5))
for i in range(2):  # Changed to 2 clusters
    plt.scatter(df_scaled[gmm_labels == i, 0], df_scaled[gmm_labels == i, 1], color=colors[i], label=labels[i])
plt.title('GMM Clustering')
plt.legend()
plt.show()

# PCA
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)

# Visualize PCA
plt.figure(figsize=(8, 5))
plt.scatter(df_pca[:, 0], df_pca[:, 1])
plt.title('PCA')
plt.show()

# t-SNE
tsne = TSNE(n_components=2)
df_tsne = tsne.fit_transform(df_scaled)

# Visualize t-SNE
plt.figure(figsize=(8, 5))
plt.scatter(df_tsne[:, 0], df_tsne[:, 1])
plt.title('t-SNE')
plt.show()

# Note: Autoencoders and SOM would require a different approach and would be more complex to implement.
```
